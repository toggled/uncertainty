```python
import numpy as np
from termcolor import colored
from torch._C import dtype
import torch
from torch.nn import Linear
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
import matplotlib.pyplot as plt
from torchsummary import summary
from torch.autograd import Variable
from keras import backend as K
import pandas as pd
import random
import json
import pickle
from itertools import islice
from numpy import load
import math
import torch_geometric
import ast
from torch_geometric.loader import DataLoader
from torch_geometric.loader import NeighborLoader
import statistics
from statistics import mean
from time import perf_counter
from torch_geometric.loader import NeighborLoader, ImbalancedSampler


LOSS_Visualization_Parameter = 50
Model_Saving_Parameter = 100

class GCN_model(torch.nn.Module):
    def __init__(self, hidden_channels):
        super(GCN_model, self).__init__()
        torch.manual_seed(1234567)
        self.conv1 = GCNConv(160, hidden_channels, bias=True)
        self.conv2 = GCNConv(hidden_channels, hidden_channels, bias=True)
        self.conv3 = GCNConv(hidden_channels, hidden_channels, bias=True)
        self.lin1 = Linear(hidden_channels, hidden_channels, bias=True)
        self.lin2 = Linear(hidden_channels, 8, bias=True)

        #self.act_f = nn.Sigmoid()
        self.act_f = nn.Softmax(dim=1)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = self.conv2(x, edge_index)
        x = x.relu()
        x = self.conv3(x, edge_index)
        x = x.relu()
        x = self.lin1(x)
        x = x.relu()
        x = self.lin2(x)
        x = self.act_f(x)
        return x

#GNN_Model = GCN(hidden_channels=100)
#print(GNN_Model)

#GNN_Model_Optimizer = torch.optim.Adam(GNN_Model.parameters(), lr=0.01, weight_decay=5e-4)


criterion = torch.nn.CrossEntropyLoss()
def loss_calculations(preds, gtruth):
    loss_per_epoch = criterion(preds, gtruth)
    return loss_per_epoch



def visualize_losses(g_losses, epoch_history, index):
    #g_losses_list = torch.stack(g_losses).cpu().detach().numpy()

    fig = plt.figure(figsize=(27,20))

    ax = plt.subplot2grid((3, 1), (0, 0), colspan=1)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title("GNN Model Loss in Epoch: " + str(epoch_history))

    ax.plot(g_losses, color='r')

    plt.savefig('/content/drive/My Drive/Node Classification on Uncertain Graph/Experiments/'+str(index)+'T_Loss_til_epoch_'+str(epoch_history)+'.png')
    #plt.savefig(str(index)+'T_Loss_til_epoch_'+str(epoch_history)+'.png')
    plt.show()


def train_step(GNN_Model, GNN_Model_Optimizer, data_loader):
    gnn_loss_list = []
    GNN_Model.train(True)
    GNN_Model_Optimizer.zero_grad()
    for batch in data_loader:

        GNN_Output = GNN_Model(batch.x, batch.edge_index)
        #print(GNN_Output[batch.train_mask].size(), batch.y[batch.train_mask].size())

        #encoded_labels = torch.zeros(batch.y.size()[0], torch.unique(batch.y).size()[0], dtype=torch.uint8)
        encoded_labels = torch.zeros(batch.y.size()[0], 8, dtype=torch.uint8)
        encoded_labels[torch.arange(batch.y.size()[0]), batch.y] = 1


        gnn_loss = loss_calculations(GNN_Output[batch.train_mask], encoded_labels[batch.train_mask].float())
        #print("gnn_loss: ", gnn_loss.tolist())
        gnn_loss_list.append(gnn_loss.tolist())
        gnn_loss.backward(retain_graph=True)

        GNN_Model_Optimizer.step()


    return mean(gnn_loss_list)


def train(GNN_Model,GNN_Model_Optimizer, EPOCHS, load_index, data_loader, index):
    gnn_training_loss_per_epoch = []

    for epoch in range(EPOCHS):
        t1 = perf_counter()
        guided_model_training_loss = train_step(GNN_Model, GNN_Model_Optimizer, data_loader)
        print(f'Epoch: {epoch+1:03d}, Model Loss: {guided_model_training_loss:.4f},   Time: {perf_counter()-t1}')

        gnn_training_loss_per_epoch.append(guided_model_training_loss)

        if (epoch + load_index + 1) % LOSS_Visualization_Parameter == 0:#and epoch > 0:
            visualize_losses(gnn_training_loss_per_epoch, epoch + load_index + 1, index)
        #if (epoch + load_index + 1) % Model_Saving_Parameter == 0 and epoch > 0:
        #  torch.save({'epoch': epoch+load_index+1, 'model_state_dict': guided_model_GCN.state_dict(), 'optimizer_state_dict': guided_optimizer.state_dict(), 'loss': guided_training_loss_per_epoch,}, "/content/drive/My Drive/" + str(Explainability_name) + " on " + str(Task_name) + "/Model/model_classifier" + str(epoch + load_index + 1)+".pt")
    #return model_loss

def gnn_test(GNN_Model, GNN_Model_Optimizer, data_x, edge_index, index, data_label, test_mask):
    test_preds = []
    test_real_labels = []
    GNN_Model.eval()
    #for batch in test_d_l:
    GNN_Output = GNN_Model(data_x, edge_index)
    GNN_Output = GNN_Output[test_mask]
    g_pred = GNN_Output.argmax(dim=1)
    test_preds.append(GNN_Output)
    r_l = data_label[test_mask]
    test_real_labels.append(r_l)
    torch.save({'model_state_dict': GNN_Model.state_dict(), 'optimizer_state_dict': GNN_Model_Optimizer.state_dict(), 'preds': test_preds, 'labels:':test_real_labels}, "/content/drive/My Drive/Node Classification on Uncertain Graph/Experiments/output_possible_world"+str(index)+".pt")


def get_ready(EPOCHS):
    #data_label = torch.load('data_label.pt')
    #train_mask = torch.load('train_mask.pt')
    #test_mask = torch.load('test_mask.pt')
    #data_x = torch.load('data_x.pt')
    data_label = torch.load('/content/drive/My Drive/Node Classification on Uncertain Graph/Dataset/author_label_scalar.pt')
    train_mask = torch.load('/content/drive/My Drive/Node Classification on Uncertain Graph/Dataset/train_mask.pt')
    test_mask = torch.load('/content/drive/My Drive/Node Classification on Uncertain Graph/Dataset/test_mask.pt')
    data_x = torch.load('/content/drive/My Drive/Node Classification on Uncertain Graph/Dataset/data_x.pt')
    data_x = data_x.type(torch.float32)

    for i in range(6,10):
        GNN_Model = GCN_model(hidden_channels=100)
        GNN_Model_Optimizer = torch.optim.Adam(GNN_Model.parameters(), lr=0.001, weight_decay=5e-4)
        edge_index = torch.load('/content/drive/My Drive/Node Classification on Uncertain Graph/Dataset/edge_index_T' + str(i) + '.pt')
        #edge_index = torch.load('edge_index_T' + str(i) + '.pt')
        #data_possible_world = torch_geometric.data.Data(x=data_x, edge_index=edge_index, y=data_label, train_mask=train_mask,
        #                                                test_mask=test_mask)
        #data_loader = NeighborLoader(data_possible_world, num_neighbors=[-1], batch_size=256,
        #                             input_nodes=torch.arange(data_possible_world.num_nodes), shuffle=False)
        #train(GNN_Model, GNN_Model_Optimizer, EPOCHS, 0, data_loader, i)
        #gnn_test(GNN_Model, GNN_Model_Optimizer, data_loader, i, data_label, test_mask)
        dataset = torch_geometric.data.Data(x=data_x, edge_index=edge_index, train_mask=train_mask, test_mask=test_mask, y=data_label)
        sampler = ImbalancedSampler(dataset)
        data_loader = NeighborLoader(dataset, batch_size=20000, num_neighbors=[-1]*1, sampler=sampler)

        train(GNN_Model, GNN_Model_Optimizer, EPOCHS, 0, data_loader, i)
        gnn_test(GNN_Model, GNN_Model_Optimizer, data_x, edge_index, i, data_label, test_mask)


get_ready(200)
```
